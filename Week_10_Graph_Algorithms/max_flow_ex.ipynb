
## Maximum Flow Problem: MATLAB Example

Given an N x N adjacency matrix representing a graph with N nodes, let each node denote a process equipment and each element of the adjacency matrix denotes the maximum flow capacity between any 2 nodes. Given a start node S and sink node T, what is the overall max amount of material that can flow from S to T?


```matlab
s = 1; t = 6;
cap = [0 16 13 0 0 0; 0 0 10 12 0 0;
       0 4 0 0 14 0;  0 0 9 0 0 20;
       0 0 0 7 0 4;   0 0 0 0 0 0];
G = digraph(cap);
plot(G,'EdgeLabel',G.Edges.Weight);
```


Solve using maxflow.


```matlab
[mf, GF] = maxflow(G,s,t);
H = plot(G,'EdgeLabel',G.Edges.Weight);
H.EdgeLabel = {};
highlight(H,GF,'EdgeColor','r','LineWidth',2);
st = GF.Edges.EndNodes;
labeledge(H,st(:,1),st(:,2),GF.Edges.Weight);
title(sprintf('Max flow: %d\n',mf));
```
