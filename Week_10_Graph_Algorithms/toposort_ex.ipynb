
## Topological Sorting: MATLAB Example

Given a set of tasks 1, 2, ..., N, how can we order them from left to right without violating any of the constraints given as: Task *i* must be performed before Task *j*. There will be K number of constraints given as a K x 2 matrix, P. Each row corresponds to two elements, i and j, which denote the constraint, interpreted as the above. If there are many possible orderings, earlier labels go first.


```matlab
N = 8; P = [3 5; 8 2; 5 1; 4 2; 4 5; 5 2; 6 2; 7 6; 4 6];
```


Create digraph of constraints.


```matlab
G = digraph(P(:,1),P(:,2),1,N);
plot(G);
```


Call topological sorting.


```matlab
[n, H] = toposort(G, 'Order','stable');
disp('Optimal ordering:')
```

```matlab
disp(n)
```

```matlab
plot(H, 'Layout', 'layered', ...
    'Direction', 'down', 'NodeLabel', cellstr(num2str(n')));
title('Suggested schedule')
```
