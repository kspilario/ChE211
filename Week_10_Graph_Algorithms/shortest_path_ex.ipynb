
## Shortest Path Problem: MATLAB Example

Given the following graph:


```matlab
s = [1 1 1 2 2 2 3 3 4 5 6 7 7 8 8 9 9];
t = [3 5 8 4 5 9 2 9 7 3 4 1 6 4 7 6 7];
w = [7 20 13 12 7 6 4 8 3 5 5 6 5 7 12 10 10];
G = digraph(s,t,w);
plot(G,'EdgeLabel',G.Edges.Weight);
```


What is the shortest path from any node S to any node T?


```matlab
S = 1; T = 6;
p = plot(G,'EdgeLabel',G.Edges.Weight);
[path,d] = shortestpath(G,S,T);
highlight(p,path,'EdgeColor','g');
title(sprintf('Shortest path length from Node %d to %d: %d\n',S, T, d));
```


Viewing the all\-pairs shortest paths:


```matlab
D = distances(G)
```
